### from cabin coulmn separate tect and numeric value

```python
import re

replace = re.compile("([a-zA-Z]+)")

df['string_cabin_list'] = df['Cabin'].str.extract(replace)
df['num_cabin_list'] = df['Cabin'].str.replace(replace, "")
```
